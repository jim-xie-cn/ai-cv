# Seaborn的使用

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from warnings import filterwarnings
```

```python
#设置画布大小
plt.style.use({'figure.figsize':(12, 6)})
#设置坐标轴和标题字体大小
sns.set(font_scale=1.5)
#关闭pandas科学计数法
pd.set_option('display.float_format', lambda x: '%.3f' % x)
plt.rcParams['font.sans-serif']=['SimHei'] #使用中文字体
```

# 格式转换 sklearn,dataframe,numpy

```python
def sklearn2df(dataset):
    df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['target'] = pd.Series(dataset.target)
    return df
def df2numpy(df):
    res = df.to_records(index=False).tolist()
    return np.array(res)
```

```python
from sklearn.feature_selection import VarianceThreshold
#读取样本数据
df = pd.read_csv("../dataset/bmi/500_Person_Gender_Height_Weight_Index.csv")
df = df.rename(columns={'Gender':'性别','Height':'身高','Weight':'体重','Index':'等级'})
#选取身高，体重为待选特征
df1 = df[['身高','体重']]
#将DataFrame格式转换为numpy数组
X = df1.to_records(index=False).tolist()
#使用方差选择法，选择方差超过500的特征
variance = VarianceThreshold(threshold=500)
newX = variance.fit_transform(X)
print("原始数据结构 :",np.array(X).shape)
print("转换后数据结构 :",newX.shape)
```

# fit,transform,fit_transform区别

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif
iris = load_iris()
X, y = iris.data, iris.target
print(X.shape)
X_new = SelectKBest(chi2, k=2).fit_transform(X, y)
X_new = SelectKBest(f_classif, k=1)
X_new.fit(X, y)
print(X_new.__dict__)
X_new = X_new.transform(X)
X_new.shape
```

# fake数据集

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100,n_features=2,centers=3, cluster_std=1.0,center_box=(-10.0,10.0),shuffle=True,random_state=1)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=50, n_features=2, n_informative=2, n_redundant=0, n_classes=2, n_clusters_per_class=1, weights=[0.5,0.5])
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.datasets import make_regression
X,y=make_regression(n_samples=1000, n_features=2,n_targets=1,noise=0,random_state=0)
print(X.shape,y.shape)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.datasets import make_friedman1
X, y = make_friedman1(n_samples=50, n_features=10, random_state=1)
print(X.shape,y.shape)
plt.scatter(X[:,0],X[:,1],c=y)
```
