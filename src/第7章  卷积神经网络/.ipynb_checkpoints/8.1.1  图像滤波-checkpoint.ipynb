```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
#读取图像，并转换为RGB格式
img= cv2.imread('./images/bridge.jpg')
image= cv2.cvtColor(img.copy(),cv2.COLOR_BGR2RGB)
#设置滤波器参数为blur算子
kernel = np.array((
        [0.0625, 0.125, 0.0625],
        [0.125, 0.25, 0.125],
        [0.0625, 0.125, 0.0625]), dtype="float32")
#进行滤波运算
dst = cv2.filter2D(image, -1, kernel)
#将原图和滤波后的图拼接到一起显示出来
all_img = np.hstack((image, dst))
plt.imshow(all_img)
```
