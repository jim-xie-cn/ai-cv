```python
# -*- coding: utf-8 -*- 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import cv2
```

```python
img= cv2.imread('./images/raw.jpg')
image= cv2.cvtColor(img.copy(),cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image.copy(),cv2.COLOR_RGB2GRAY)
edge = cv2.Canny(gray.copy(),128,255)
ret, binary = cv2.threshold(gray.copy(), 80, 255, cv2.THRESH_BINARY)
binary = cv2.bitwise_not(binary)
blur = cv2.GaussianBlur(gray.copy(),(5,5),3)

fig,ax = plt.subplots(2,2,figsize=(30,10))
plt.subplot(2,2,1)
plt.imshow(image)
plt.subplot(2,2,2)
plt.imshow(gray,cmap='gray')
plt.subplot(2,2,3)
plt.imshow(binary,cmap=plt.get_cmap('binary'))
plt.subplot(2,2,4)
plt.imshow(edge,cmap='gray')
binary
```

```python
img= cv2.imread('./images/raw.jpg')
img=cv2.resize(img,(12,16))
gray=cv2.cvtColor(img.copy(),cv2.COLOR_BGR2GRAY)
cv2.imwrite('./images/gray.jpg',img)
ret, binary = cv2.threshold(gray.copy(), 80, 255, cv2.THRESH_BINARY)
binary = cv2.bitwise_not(binary)
cv2.imwrite('./images/bin.jpg',binary)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imwrite('./images/hsv.jpg',hsv)
```

```python
data=[]
for y in range(len(gray)):
    tmp = {}
    for x in range(len(gray[y])):
        d = gray[y][x]
        tmp[str(x)]='['+str(d)+']'
    data.append(tmp)
df = pd.DataFrame(data)
df.to_csv("./output/gray.csv")
```

```python
data=[]
for y in range(len(img)):
    tmp = {}
    for x in range(len(img[y])):
        d = img[y][x]
        tmp[str(x)]=str(d)
    data.append(tmp)
df = pd.DataFrame(data)
df.to_csv("./output/img.csv")
```

```python
data=[]
for y in range(len(binary)):
    tmp = {}
    for x in range(len(binary[y])):
        d = binary[y][x]
        tmp[str(x)]=str(d)
    data.append(tmp)
df = pd.DataFrame(data)
df.to_csv("./output/binary.csv")
```

```python
data=[]
for y in range(len(hsv)):
    tmp = {}
    for x in range(len(hsv[y])):
        d = hsv[y][x]
        tmp[str(x)]=str(d)
    data.append(tmp)
df = pd.DataFrame(data)
df.to_csv("./output/hsv.csv")
```
