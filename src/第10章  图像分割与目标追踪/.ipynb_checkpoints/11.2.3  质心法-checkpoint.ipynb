导入库

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #在统计图上显示中文
```

```python
net = cv2.dnn.readNetFromDarknet('./models/yolov3-tiny.cfg','./models/yolov3-tiny.weights')
```

对图像进行目标检测

```python
# 加载图片、转为blob格式、送入网络输入层
img = cv2.imread('./images/ball_4.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #转换为RGB格式
blobImg = cv2.dnn.blobFromImage(img, 1.0/255.0, (416, 416), None, True, False)  ## net需要的输入是blob格式的，用blobFromImage这个函数来转格式
net.setInput(blobImg)  ## 调用setInput函数将图片送入输入层
```

获取网络输出层信息

```python
outInfo = net.getUnconnectedOutLayersNames() 
layerOutputs = net.forward(outInfo)
```

```python
(H, W) = img.shape[:2]
boxes = [] # 所有边界框（各层结果放一起）
confidences = [] # 所有置信度
classIDs = [] # 所有分类ID
for out in layerOutputs:  # 各个输出层
    for detection in out:  # 各个框框
        # 拿到置信度
        scores = detection[5:]  # 各个类别的置信度
        classID = np.argmax(scores)  # 最高置信度的id即为分类id
        confidence = scores[classID]  # 拿到置信度
        if confidence > 0.3:
            box = detection[0:4] * np.array([W, H, W, H])  # 将边界框放会图片尺寸
            (centerX, centerY, width, height) = box.astype("int")
            x = int(centerX - (width / 2))
            y = int(centerY - (height / 2))
            boxes.append([x, y, int(width), int(height)])
            confidences.append(float(confidence))
            classIDs.append(classID)
            print(classID,float(confidence),[x, y, int(width), int(height)])
```

```python
# 得到labels列表
with open("./models/coco.names", 'rt') as f:
    labels = f.read().rstrip('\n').split('\n')
```

```python
idxs = cv2.dnn.NMSBoxes(boxes, confidences, 0.3, 0.1) 
```

```python
COLORS = np.random.randint(0, 255, size=(len(labels), 3), dtype="uint8") 
mask = img.copy()
if len(idxs) > 0:
    for i in idxs.flatten():
        (x, y) = (boxes[i][0], boxes[i][1])
        (w, h) = (boxes[i][2], boxes[i][3])

        color = [int(c+255) for c in COLORS[classIDs[i]]]
        cv2.rectangle(mask, (x, y), (x+w, y+h), color, 5) 
        text = "{}: {:.2f}".format(labels[classIDs[i]], confidences[i])
        cv2.putText(mask, text, (x-20, y-20), cv2.FONT_HERSHEY_SIMPLEX, 0.75, color, 1)
plt.imshow(mask)
```
